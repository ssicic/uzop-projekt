### 1) Priprema podataka

```python
#metabolicki geni
df = pd.ExcelFile('C:/Users/Sara/Documents/My Fax/UZOP/uzop-projekt/files/metabolicki_geni.xls').parse('All metabolic genes') #you could add index_col=0 if there's an index
gene_symbols=[]
for cell in df['Gene Symbol']:
    gene_symbols.append(cell)
gene_set=set(gene_symbols)
```

```python
#anotacije
annot = read_gtf("C:/Users/Sara/Documents/My Fax/UZOP/uzop-projekt/files/anotacije_gena.gtf")
gene_dict_id = dict()
gene_dict_len = dict()

gene_set_2 = set()
for i in range (0, len(annot)):
    if annot["gene_name"][i].upper() in gene_set:
        if ((annot["end"][i]-annot["start"][i])!=0):
            gene_set_2.add(annot["gene_name"][i].upper())
            gene_dict_id[annot["gene_id"][i].upper()]=annot["gene_name"][i]
            gene_dict_len[annot["gene_name"][i].upper()]=annot["end"][i]-annot["start"][i]
```

```python
#icgc ucitavanje
file_icgc = open('C:/Users/Sara/Documents/My Fax/UZOP/uzop-projekt/files/ICGC_exp_seq.tsv')
icgc_tsv = csv.reader(file_icgc, delimiter="\t")
n = 0
icgc_dict = dict()
rCounts = dict()
for row in icgc_tsv:
    #print(row)
    n = n + 1
    if (n==1000000):
        break
    if (row[7] in gene_set_2):
        icgc_dict[(row[7], row[0])]=int(row[9])/int(gene_dict_len[row[7]])
        if (int(row[9])>0):
            if row[1] in rCounts:
                rCounts[row[0]]+=int(row[9])/int(gene_dict_len[row[7]])
            else:
                rCounts[row[0]]=int(row[9])/int(gene_dict_len[row[7]])
for key in icgc_dict:
    icgc_dict[key]=(icgc_dict[key]/rCounts[key[1]])/1000000
```

```python
#tcga ucitavanje

dir_tcga = os.listdir('C:/Users/Sara/Documents/My Fax/UZOP/uzop-projekt/files/tcga')
n=0
tcga_dict = dict()
rCounts2 = dict()
for dir in dir_tcga:
    n += 1
    dirName = 'C:/Users/Sara/Documents/My Fax/UZOP/uzop-projekt/files/tcga/'+dir
    for f in listdir(dirName):
        
        if (f.endswith('gz')):
            file = dirName+'/'+ f
            patientName = f[0:len(f)-16]
            if (isfile(file)):
                with gzip.open(file, "r") as f_in:
                    content = f_in.read()
                    content2=[]
                    content2 = str(content.decode("utf-8")).split('\n')
                    for c in content2:
                        sample = c.split('\t')
                        if sample[0] in gene_dict_id.keys():
                            
                            tcga_dict[gene_dict_id[sample[0]],patientName]=int(sample[1])/int(gene_dict_len[gene_dict_id[sample[0]]])
                            if (int(sample[1])>0):
                                if patientName in rCounts:
                                    rCounts2[patientName]+=int(sample[1])/int(gene_dict_len[gene_dict_id[sample[0]]])
                                else:
                                    rCounts2[patientName]=int(sample[1])/int(gene_dict_len[gene_dict_id[sample[0]]])
    if (n==20):
        break
for key2 in tcga_dict:
    tcga_dict[key2]=(tcga_dict[key2]/rCounts2[key2[1]])/1000000
```

```python
#tablica
#TO DO: dict u tablice, mergeat tablice u jednu, Pycombat

l1 = list(tcga_dict.keys())
rows = []
columns = []
n = 0
l1 = list(icgc_dict.keys())
print(len(l1))
all_dict = icgc_dict
for tcga_entry in tcga_dict:
    all_dict.update({tcga_entry : tcga_dict[tcga_entry]})
for a in all_dict:
    rows.append(a[0])
    columns.append(a[1])
print(len(rows))
#df = pd.DataFrame(all_dict, index=columns)
```

### 2) Odabir bitnih gena i klasteriranje

```python
#MAD
#>0.5
import scipy as sp

all_dict = tcga_dict.copy()
filter_dict = all_dict.copy()
new_dict = dict()
patientSet = set()

remaining_genes = set()
result_genes = set()

for x in all_dict:
    remaining_genes.add(x[0])

for entry in all_dict:
    values = []
    if (entry[0] in remaining_genes):
        for e in all_dict:
            if e[0]==entry[0]:
                values.append(all_dict[e])
        remaining_genes.remove(entry[0])
    print(sp.stats.median_abs_deviation(values))
    if (sp.stats.median_abs_deviation(values)>0.5):
        result_genes.add(entry[0])
print(result_genes)
```
